```python
data.drop(['Unnamed: 0', 'Date'], axis=1, inplace=True)
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
```

```python
data['Location'] = encoder.fit_transform(data['Location'])

e_o = dict(zip(encoder.transform(encoder.classes_),
               encoder.classes_))

for i, j in e_o.items():
  print(f'{i} - {j}')
```

```python
data['Tomorrow_Rain'].value_counts()
```

```python
x = data.drop('Tomorrow_Rain', axis=1)
y = data['Tomorrow_Rain']
```

```python
x
```

```python
y
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
```

```python
x_oversampled, y_oversampled = smote.fit_resample(x, y)

df = pd.concat([x_oversampled, y_oversampled], axis=1)

df
```

```python
df['Tomorrow_Rain'].value_counts()
```

```python
from pycaret.classification import *
```

```python
s = setup(data=df, target='Tomorrow_Rain', session_id=123)

best_models = compare_models()
```

```python
model = create_model('rf')

model
```

```python
model.score(x, y)
```

```python
import numpy as np

features = np.array([[loc, temp, min, max, wind_speed, wind_dir, humid, press, cloud, rain]])

predictions = model.predict(features)[0]

if predictions == 1:
  print('Tomorrow will rain')
else:
  print('Tomorrow will not rain')
```
